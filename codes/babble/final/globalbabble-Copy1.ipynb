```python
import os
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import numpy as np
outputpath = '/media/hd2/chaitanya/Deep_Learning/project/Spectrograms/Noisy_TCDTIMIT/Babble'
NoiseBabble = []
for dirpath,dirnames,filenames in os.walk('/media/hd2/chaitanya/Deep_Learning/project/TCDTIMIT/Noisy_TCDTIMIT/Babble/'):
    print("present dir: ",dirpath)
    structure = os.path.join(outputpath, dirpath[len('/media/hd2/chaitanya/Deep_Learning/project/TCDTIMIT/Noisy_TCDTIMIT/Babble/'):])
    for filename in filenames:
        if filename.endswith(".wav"):
            print("wavfile: ",filename)
            sample_rate, samples = wavfile.read(os.path.join(dirpath,filename))
            f, t, Zxx = signal.stft(samples, sample_rate,nperseg=256,nfft=256)
            T = len(t)//16
            #print(Zxx.shape)
            for i in range(0,(T)*16,16):
                NoiseBabble.append(np.log(np.abs(Zxx[:,i:i+16])+1e-8))
            


```
