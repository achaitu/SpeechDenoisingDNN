```python
import os
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import numpy as np
NoiseCafe = []
for dirpath,dirnames,filenames in os.walk('/media/hd2/chaitanya/Deep_Learning/project/TCDTIMIT/Noisy_TCDTIMIT/Cafe/-5/'):
    print("present dir: ",dirpath)
    for filename in filenames:
        if filename.endswith(".wav"):
            print("wavfile: ",filename)
            sample_rate, samples = wavfile.read(os.path.join(dirpath,filename))
            f, t, Zxx = signal.stft(samples, sample_rate,nperseg=256,noverlap=192,nfft=256)
            T = len(t)//16
            #print(Zxx.shape)
            for i in range(0,(T)*16,16):
                NoiseCafe.append(np.log(np.abs(Zxx[:,i:i+16])+1e-8))
            


```
